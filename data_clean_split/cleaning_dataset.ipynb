## Dataset columns

- **Name** (str) - Name of the passenger
- **Pclass** (int) - Ticket class
- **Sex** (str) - Sex of the passenger
- **Age** (float) - Age in years
- **SibSp** (int) - Number of siblings and spouses aboard
- **Parch** (int) - Number of parents and children aboard
- **Ticket** (str) - Ticket number
- **Fare** (float) - Passenger fare
- **Cabin** (str) - Cabin number
- **Embarked** (str) - Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

titanic = pd.read_csv('../resources/titanic.csv')
titanic.head()
```

### Cleaning data

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(),inplace=True) #Replace null age  with mean age 
titanic.head(7)
```

```python
for i,col in enumerate(['SibSp','Parch']):
    plt.figure(i)
    sns.catplot(x=col,y='Survived',data=titanic,kind='point',aspect=2,)
```

### Combine SibSp,Parch

While Number of Parch and SibSp increasing, survival rate is decresing .So both of them are same trajectory so we can combine together.

```python
titanic['Family_cnt'] = titanic['SibSp'] + titanic['Parch']
```

#### Drop unnecessary variables

```python
titanic.drop(['SibSp','Parch','PassengerId'],inplace=True,axis=1)
titanic.head()
```

### Cleaning Categorical variables

```python
titanic.isnull().sum()
```

```python
titanic.groupby(titanic['Cabin'].isnull())['Survived'].mean()
```

```python
titanic['Cabin_ind']=np.where(titanic['Cabin'].isnull(),0,1)
titanic.head()
```

### convert sex to neumeric

```python
gender_num={"male":0,"female":1}
titanic['Sex']=titanic['Sex'].map(gender_num)
```

#### Drop unnecessary variables

```python
titanic.drop(['Cabin', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
titanic.head()
```

### Write out cleaned data

```python
titanic.to_csv('../resources/titanic_cleaned.csv',index=False)
```
